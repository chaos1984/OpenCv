```python
import numpy as np
import cv2
import shutil
import os
```

```python
def cv_show(name,img):
    cv2.imshow(name,img) 
    cv2.waitKey(0) 
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('68.jpg')
cv_show('img',img)
```

```python
bac = cv2.imread('46.jpg')
bac  = cv2.pyrDown(bac.copy())
bac  = cv2.pyrDown(bac)
cv_show('bac',img)
```

```python
HSV = cv2.cvtColor(bac.copy(), cv2.COLOR_BGR2HSV)  #RGB  HSV

H, S, V = cv2.split(HSV)    # HSV 
bac_s = S
bac_s = cv2.bilateralFilter(bac_s, d=0, sigmaColor=100, sigmaSpace=15)
# bac_s = cv2.threshold(bac_s, 60, 255, cv2.THRESH_BINARY)
# bac_s = cv2.threshold(bac_s , 127, 255, cv2.THRESH_BINARY)
cv_show('bac_s',bac_s)
```

```python
frame = cv2.imread('68.jpg')
frame  = cv2.pyrDown(frame.copy())
frame  = cv2.pyrDown(frame)
cv_show('frame',frame)
```

```python
HSV = cv2.cvtColor(frame.copy(), cv2.COLOR_BGR2HSV)  #RGB  HSV

H, S, V = cv2.split(HSV)    # HSV 
frame_s = S
frame_s = cv2.bilateralFilter(frame_s, d=0, sigmaColor=100, sigmaSpace=15)
# frame_s = cv2.threshold(frame_s, 60, 255, cv2.THRESH_BINARY)
# frame_s = cv2.threshold(frame_s , 127, 255, cv2.THRESH_BINARY)
cv_show('frame_s',frame_s)
```

```python
from skimage.measure import compare_ssim as ssim
score,gra, diff = ssim(frame_s,bac_s,  full=True,gradient=True,gaussian_weights=True,win_size=3)

# print("SSIM: {}".format(score))
diff = (diff*255).astype('uint8')
# cv2.imshow('diff',diff)
# diff = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
sub = cv2.absdiff(frame_s,bac_s)
sub1 = cv2.bitwise_xor(frame_s,bac_s)
res = np.hstack((bac_s,frame_s,diff,sub,sub1))
cv_show('diff',res)
```
